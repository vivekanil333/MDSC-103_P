```python
import numpy as np
from scipy.optimize import minimize

# Define the function to minimize
def fun_to_minimize(x):
    x1, x2 = x
    return 5*x1**2 + x2**2 + 2*x1*x2

# Constraint function: x1*x2 - 10 = 0
def constraint_eq(x):
    x1, x2 = x
    return x1 * x2 - 10

# Initial guess for the minimum (starting point)
x0 = [0, 0]

# Set up the constraint object
constraint = {'type': 'eq', 'fun': constraint_eq}

# Use minimize to find the local minimum with the constraint
result = minimize(fun_to_minimize, x0, constraints=constraint, method='SLSQP')

# Extract the minimized values of x1 and x2
x1_min = result.x[0]
x2_min = result.x[1]
min_value = result.fun

print("Minimum point (x1, x2):", (x1_min, x2_min))
print("Minimum value of the function:", min_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def fun_to_minimize(x):
    x1, x2 = x
    return 5*x1**2 + x2**2 + 2*x1*x2

# Constraint function: x1*x2 - 10 = 0
def constraint_eq(x):
    x1, x2 = x
    return x1 * x2 - 10

# Generate data for plotting the function
x1_vals = np.linspace(-10, 10, 100)
x2_vals = np.linspace(-10, 10, 100)
x1_grid, x2_grid = np.meshgrid(x1_vals, x2_vals)
z = 5*x1_grid**2 + x2_grid**2 + 2*x1_grid*x2_grid

# Create the contour plot for the function
plt.contourf(x1_grid, x2_grid, z, levels=50, cmap='viridis')

# Plot the constraint line x1*x2 = 10
constraint_line_x2 = np.linspace(-10, 10, 100)
constraint_line_x1 = 10 / constraint_line_x2
plt.plot(constraint_line_x1, constraint_line_x2, 'r--', label='Constraint: x1*x2=10')

# Add a legend and labels
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Function Contour Plot with Constraint')
plt.legend()

# Find the minimum point and add it to the plot
from scipy.optimize import minimize
result = minimize(fun_to_minimize, [0, 0], constraints={'type': 'eq', 'fun': constraint_eq})
x1_min = result.x[0]
x2_min = result.x[1]
min_value = result.fun
plt.scatter(x1_min, x2_min, color='red', marker='o', label='Minimum (with constraint)')

# Show the plot
plt.legend()
plt.colorbar(label='Function Value')
plt.show()

print("Minimum point (x1, x2):", (x1_min, x2_min))
print("Minimum value of the function:", min_value)
```
